# Impute functinos

```python
def get_defaults(features, labels, columns):
    """Default values for nans
    
    For subtask1's features nans: fill it with the average of all the corresponding negative-labeled features
    (e.g.) nans of 'AST' will be filled with the average of all the AST's negative-labeled features
    For subtask2's features nans: fill it with the same way as for subtask1's features
    For subtask3's features nans: fill it with (fill it with the values that the same-age(or closest) person has) or (the average)
                                                        In this function, default value is the average. The other method is defined in get_12rows function
    For other feartures nans: fill it with the average
    
    Args:
        features(df): 
        labels(df):
        columns(list): features' columns except 'pid' and 'TIme' since they cannot be nan
        
    Returns:
        defaults(dict): keys are column name, values are default values for that keys
    
    """
    import numpy as np
    defaults = dict()
    for column in columns:
        col_name = 'LABEL_' + column
        tmp = 0.
        print(column)
        if column in SUBTASK1:
            negative_labels_pids = labels.loc[labels[col_name] == 0.0, 'pid']
            denom = len(negative_labels_pids)
            for pid in negative_labels_pids:
                # if all are null don't count it
                if features.loc[features['pid'] == pid, column].isnull().all():
                    denom -= 1
                else:
                    tmp += features.loc[features['pid'] == pid, column].mean(skipna=True)
            defaults[column] = tmp / denom
            
        elif column in SUBTASK2:
            negative_labels_pids = labels.loc[labels[col_name] == 0.0, 'pid']
            denom = len(negative_labels_pids)
            for pid in negative_labels_pids:
                # if all are null don't count it
                if feature.loc[features['pid'] == pid, column].isnull().all():
                    denom -= 1
                else:
                    tmp += features.loc[features['pid'] == pid, column].mean(skipna=True)
            defaults[column] = tmp / denom
            
        elif column in SUBTASK3:
            defaults[column] = features[column].mean(skipna=True)
            
        else:
            defaults[column] = features[column].mean(skipna=True)
            
    return defaults
```

```python
defaults = get_defaults(df_train_features, df_train_labels, columns)
```

```python
def get_12rows(features, rows, column, pid):
    """Get a person's data(12 x 1 dim data)
    
    fill nan.
    SUBTASK3's nans will be filled differently.
        first choice is to fill nans with the average of closest-aged  people's data
        second choice is to fill nans with defalut value got in get_defaults function
    
    Args:
        features(df): the whole dataframe
        rows(df): 12-rowed dataframe that has the same pid (pid, column)
        column: a feature
        pid(int): 
        
    Returns:
        rows(df): 12-rowed dataframe whose nans are filled in this function
        
    """
    import numpy as np
    
    if rows.any():
        rows.fillna(rows.mean(), inplace=True)

    else:
        if column in SUBTASK3:
            age = features.loc[features['pid']==pid, 'Age'].iat[0]
            min_age_distance = (features.loc[features['pid']!=pid, 'Age']- age).abs().min()
            df = df_train_features[df_train_features['pid']!=pid] # df_train features without current pid
            closest_age_pids =  df.loc[(df['Age']-age).abs() == min_age_distance, 'pid'].unique()# select indexes s.t. its ages are close
            default = features.loc[features['pid'].isin(closest_age_pids), 'RRate'].mean()
            rows.fillna(default, inplace=True)
            if not rows.any():
                rows.fillna(defaults[column], inplace=True)
            
        else:
            rows.fillna(defaults[column], inplace=True)
    
    return rows
```

```python
def fill_nan(features, columns, pids):
    """imputing variable's strategy
    
    Args:
        features(df): the whole dataframe
        columns(list): features' columns except 'pid' and 'TIme' since they cannot be nan
        pids(list): pids. they are different from each other
        
    Returns:
        features(df): the whole dataframe whose nans are filled completly using 
    
    """
    for column in columns:
        print(column)
        for pid in pids:
            rows = features.loc[features['pid']==pid, column] # 12 data (for a person)
            features.loc[features['pid']==pid, column] = get_12rows(features, rows, column, pid)
    
    return features
```

# Training fuction

```python
def model_fit(train_X, train_y, val_X, val_y, tasks, feature_selection, model):
    """
    
    Apply feature selection and training for each task.
    
    Args:
        X(df):
        y(df):
        feature_selection(func): 
        model(func):
        
    Returns:
        results(dict of dict): {task name: {hyper para11: score11, hyper para12: score12}, ..., {hyper para21: score21, }}
        transformers(dict): {task name: feature selection model, ... }
        
    """
    results = dict()
    models = dict()
    transformers = dict()
    for task in tasks:
        print(task)
        train_X_new, transformers[task] = feature_selection(train_X, train_y['LABEL_'+task])
        val_X_new = transformers[task].transform(val_X)
        results[task] = model(train_X_new, train_y['LABEL_'+task], val_X_new, val_y['LABEL_'+task])
    
    return results, transformers
```

```python
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import metrics
from sklearn.metrics import roc_auc_score

def model1(train_X, train_y, val_X, val_y):
    """
    Train a model and evaluate it using ROC_AUC. 
    The activateion function is sigmoid. 
    
    Args:
        X(df) : Features that are useful for prediction
        y(Series) : The answers 
        
    Returns:
        scores(dict) : {task: score, ...}
    
    """
    clf = LinearSVC(random_state=0, class_weight='balanced', dual=False)
    #clf = LogisticRegression(random_state=0)
    
    C_s = [0.01, 0.1, 1, 5]
    
    scores = dict()
    for c in C_s:
        clf.C = c
        model = make_pipeline(PolynomialFeatures(degree=3), clf)
        model.fit(train_X, train_y)
        pred_y = model.predict(val_X)
        scores[c] = roc_auc_score(val_y, pred_y)
    
    return scores
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import SelectFromModel

def feature_selection3(X, y):
    """Feature selection for subtask3
    
    Automatically select features by performing Lasso(squared loss with l1 penalty) on train dataset
    
    Args:
        X(df): Features to make a prediction
        y(Series): The answers 
        
    Returns:
        X_new(df): selected features
        model: how to select features from its original (to perform the same feature selection method on test set)
        
    """
    lasso = Lasso(alpha=0.01).fit(X, y)
    model = SelectFromModel(lasso, prefit=True, max_features=5)
    X_new = model.transform(X)
    return X_new, model
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

def model3(train_X, train_y, val_X, val_y):
    """Model for subtask3
    
    Train a model and evaluate it using R2 score. 
    
    Args:
        X(df): Features to make a prediction
        y(Series): The answers 
        
    Returns:
        scores(dict): {task: score, ...}

    """
    reg = Ridge(random_state=0)
    alphas = [0, 0.01, 0.1, 1, 10]
    
    scores = dict()
    for alpha in alphas:
        reg.alpha = alpha
        model = make_pipeline(PolynomialFeatures(degree=4), reg)
        model.fit(train_X, train_y)
        pred_y = model.predict(val_X)
        scores[alpha] = r2_score(val_y, pred_y)
        
    return scores
```

# Helper functions

```python
import numpy as np

def prediction(df, models, tasks, transformers):
    """Get results for predictions
    
    This function is to make predictions for given df with given models. 
    The results will be returned in a combined form of the predictions for all the tasks
    
    Args:
        df(df): df that you want to make predictions of
        models: models that are used to make predictions on
    
    Returns:
        df_total_result(df): Combined results
        
    TODO:
        perform the same feature selection
    
    """
    df_pid = df[['pid']].drop_duplicates(ignore_index=True)
    df_total_result = pd.DataFrame(data=df_pid.reindex(), columns=['pid'])
    df = df.drop(columns=['pid'])
    tmp_sum= 0
    
    for task in tasks:
        predictions = models[task].predict(transformers[task].transform(df))
        predictions = predictions.reshape(predictions.shape[0], 1)
   
        result_for_a_task = list()
        for i, prediction in enumerate(predictions):
            if i % 12 == 11:
                tmp_sum /= 12.
                result_for_a_task.append(tmp_sum)
                tmp_sum = 0
            tmp_sum += prediction
        else:
            if i % 12 != 11:
                tmp_sum /= (i % 12 + 1)
            print(f'done: {task}')
        
        record = pd.DataFrame(data=result_for_a_task , columns=['LABEL_' + task])
        df_total_result = pd.concat([df_total_result, record], axis=1)

    return df_total_result
```

```python
def output_to_csv(df, name):
    """Output df as csv
    
    Args:
        df(df): df that you want to output
        name(str): file name without its extension
    
    """
    file_name = name + '.zip'
    df.to_csv(file_name, index=False, float_format='%.3f', compression='zip')
```
